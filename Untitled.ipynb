```python
from sympy import *
```

```python
w,F = symbols('omega F')
fun = F*2/3*w**3 - 2/7 * w**7
wmax = solve(fun.diff(w),w)[2]
display(wmax)
display(fun.subs(w,wmax))
```

```python
F_value =1e-4
plot(fun.subs(F,F_value),(w,0,F_value**0.25))
```

```python
plot(wmax,fun.subs(w,wmax),(F,1e-10,1),yscale="log",xscale="log")
```
