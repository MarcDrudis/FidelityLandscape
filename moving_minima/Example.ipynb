```python
from sympy import symbols,sin,plot,diff,solve
from sympy_slider import slider
import numpy as np
import plotly.express as px
%matplotlib inline
%matplotlib widget
%config InlineBackend.figure_formats = ['svg']
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [3, 2]
```

```python
x,t = symbols("x t")
```

```python
f = t * (sin(x)**3+1)/2 +(1-t)*sin(x)**2+t*(sin(x)+1)/3
df=diff(f,x)
ddf=diff(df,x)
```

```python
solutions = solve(df,x)
display(solutions)
display(f)
```

```python
plot(solutions[1],(t,0,1))
```

```python
from symplot import prepare_sliders
import matplotlib.pyplot as plt
import mpl_interactions.ipyplot as iplt
import numpy as np
import sympy as sp
np_expression, sliders = prepare_sliders(f, plot_symbol=x)
sliders.set_ranges(
    t=(0,1,1e-2)
)
sliders.set_values(t=1e-2)

plot_domain = np.linspace(-2*np.pi, 2*np.pi, 1_000)
fig, ax = plt.subplots(figsize=(3, 2))
controls = iplt.plot(
    plot_domain,
    np_expression,
    **sliders,
    ylim="auto",
)
ax.vlines(x=0,ymin=0,ymax=1)
ax.set_xlabel("$x$")
ax.set_ylabel(f"${sp.latex(f)}$");
```

```python
np_expression, sliders = prepare_sliders(df, plot_symbol=x)
sliders.set_ranges(
    t=(0, 1.1,1e-3),
)
sliders.set_values(t=1e-3)

plot_domain = np.linspace(-3,0.1, 1_000)
fig, ax = plt.subplots(figsize=(3, 2))
controls = iplt.plot(
    plot_domain,
    np_expression,
    **sliders,
    ylim="auto",
)
ax.hlines(y=0,xmin=-3,xmax=0.1)
ax.set_xlim((-0.4,0.1))
ax.set_xlabel("$x$")
ax.set_ylabel(f"${sp.latex(df)}$");
```

```python
from symplot import prepare_sliders
import matplotlib.pyplot as plt
import mpl_interactions.ipyplot as iplt
import numpy as np
import sympy as sp
np_expression, sliders = prepare_sliders(f, plot_symbol=x)
np_expression1, sliders1 = prepare_sliders(df, plot_symbol=x)
np_expression2, sliders2 = prepare_sliders(ddf, plot_symbol=x)
sliders.set_ranges(
    t=(0,1,1e-2)
)
sliders.set_values(t=1e-2)

plot_domain = np.linspace(-2*np.pi, 2*np.pi, 1_000)
fig, ax = plt.subplots(figsize=(3, 2))
controls = iplt.plot(
    plot_domain,
    np_expression,
    **sliders,
    ylim="auto",
)
controls = iplt.plot(
    plot_domain,
    np_expression1,
    **sliders,
    ylim="auto",
)
controls = iplt.plot(
    plot_domain,
    np_expression2,
    **sliders,
    ylim="auto",
)
ax.vlines(x=0,ymin=0,ymax=1)
ax.hlines(y=0,xmin=-3,xmax=0.1)
ax.set_xlim((-0.4,0.1))

ax.set_xlabel("$x$")
ax.set_ylabel(f"${sp.latex(f)}$");
```
